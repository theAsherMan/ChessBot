```python
from ChessBots import ZeroDepthEngine
from NN.Devices import best_device
from Chess_Utils import Board
```

```python
def displayBoard(board:Board):
  display(SVG(chess.svg.board(board=board, size=650)))
```

```python
engine = ZeroDepthEngine('/saves/ZeroDepthEngine')
while True:
    for ii in range(100):
        engine.new_game()
        game = Board()
        while not game.is_game_over():
            if ii==0:
                displayBoard(game)
                sleep(1)
            move = engine.makeMove(game)
            game.push(move)
        engine.train(winner=game.outcome().winner)
```
