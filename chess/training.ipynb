```python
from ChessBots import ZeroDepthEngine
from NN.Devices import best_device
from Chess_Utils import Board
```

```python
def displayBoard(board:Board):
  display(SVG(chess.svg.board(board=board, size=650)))
```

```python
engine = ZeroDepthEngine('/saves/ZeroDepthEngine')
while True:
    for ii in range(100):
        boards = []
        moves = []
        engine.new_game()
        game = Board()
        while not game.is_game_over():
            if ii==0:
                clear_output(wait=True)
                displayBoard(game)
                #sleep(1)
            boards.append(game.copy())
            move = engine.makeMove(game)
            moves.append(move)
            game.push(move)
        print('game over')
        print(f'outcome:{game.outcome()}')
        engine.train(boards=boards, moves=moves, outcome=game.outcome())
```
